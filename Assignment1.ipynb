## 6. Which school is the closest to a residence ID = 3115? Hint: Use Euclidean distance using the lat and long.

```python
points = []

task6_data_3115 = dataset[(dataset['residence_id'] == 3115)]
task6_data_3115[['residence_id','school_id','lat','lon','currentLocationID','currentLocationType']].head(5)
lat_3115 = task6_data_3115.lat.values[0]
lon_3115 = task6_data_3115.lon.values[0]

task6_data_school = dataset[(dataset['currentLocationType'] == 'school')].drop_duplicates(subset = ['currentLocationID'])
task6_data_school[['residence_id','school_id','lat','lon','currentLocationID','currentLocationType']]

pass_to_loop = task6_data_school[['lat','lon','school_id']]
fixed_point = (lat_3115,lon_3115)


for index, row in pass_to_loop.iterrows():
#     print(row['lat'], row['lon'])
    change_point = (row['lat'],row['lon'])
    points.append([int(row['school_id']),row['lat'],row['lon'],distance.euclidean(fixed_point,change_point)])
loop_out = pd.DataFrame(points, columns=["school_id", "latitude","longitude","closest_distance"])
loop_out[(loop_out['closest_distance'] == loop_out.closest_distance.min())]




# print(type(fixed_point))
# d = distance.euclidean(p1, p2)
```
